```python
plt.figure(figsize=(20, 15))
plt.subplot(2,2,1)
df['Ship Mode'].value_counts(normalize=True).plot.pie()
plt.subplot(2,2,2)
df['Segment'].value_counts(normalize=True).plot.pie()
plt.subplot(2,2,3)
df['Region'].value_counts(normalize=True).plot.pie()
plt.subplot(2,2,4)
df['Category'].value_counts(normalize=True).plot.pie()
```

```python
df1=df.groupby('Segment')['Quantity'].sum().reset_index()
df1
```

```python
plt.figure(figsize=(5,5))
plt.pie(df1['Quantity'], labels=df['Segment'].unique(), autopct='%1.1f%%')
plt.title("Quantities ordered by each segment", size=25)
plt.show()
```

```python
df2=df.groupby('Region')['Quantity'].sum().reset_index()
df2
```

```python
plt.figure(figsize=(5,5))
plt.pie(df2['Quantity'], labels=df['Region'].unique(), autopct='%1.1f%%')
plt.title("Quantities ordered by each Region", size=25)
plt.show()
```

```python
df3=df.groupby('Category')['Quantity'].sum().reset_index()
df3
```

```python
plt.figure(figsize=(5,5))
plt.pie(df3['Quantity'], labels=df['Category'].unique(), autopct='%1.1f%%')
plt.title("Quantities sold in each Category", size=25)
plt.show()
```

```python
df4=df.groupby('Segment')['Profit'].sum().reset_index()
df4
```

```python
plt.figure(figsize=(5,5))
plt.pie(df4['Profit'], labels=df['Segment'].unique(), autopct='%1.1f%%')
plt.title("Profits in each segment", size=25)
plt.show()
```

```python
df5=df.groupby('Region')['Profit'].sum().reset_index()
print(df5)
plt.figure(figsize=(5,5))
plt.pie(df5['Profit'], labels=df['Region'].unique(), autopct='%1.1f%%')
plt.title("Profits in each Region", size=25)
plt.show()
```

```python
df6=df.groupby('Category')['Profit'].sum().reset_index()
print(df6)
plt.figure(figsize=(5,5))
plt.pie(df6['Profit'], labels=df['Category'].unique(), autopct='%1.1f%%')
plt.title("Profits in each Category", size=25)
plt.show()
```

```python
sns.countplot("Ship Mode", data=df)
```

```python
sns.catplot("State", data=df, kind='count', aspect=2, height=6)
plt.xticks(rotation=90)
plt.show()
```

```python
df7=df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)
print(df7)
sns.catplot("Sub-Category", "Profit", data=df7, kind='bar', aspect=2, height=6)
plt.xticks(rotation=90)
plt.show()
```

```python
df8=df.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)
print(df8.head(10))
sns.catplot("City", "Profit", data=df8.head(10), kind='bar', aspect=2, height=6)
plt.xticks(rotation=90)
plt.show()
```
